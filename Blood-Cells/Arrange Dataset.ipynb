```python
import pandas as pd
import os
import os.path
import numpy as np
import shutil
```

```python
dataset = pd.read_csv('labels.csv')
```

```python
index = dataset.iloc[:,1:2].values
```

```python
name = dataset.iloc[:, 2:].values
```

```python
paths = {
    'EOSINOPHIL': os.path.join('EOSINOPHIL'),
    'NEUTROPHIL': os.path.join('NEUTROPHIL'),
    'MONOCYTE': os.path.join('MONOCYTE'),
    'LYMPHOCYTE': os.path.join('LYMPHOCYTE'),
    'Annotations': os.path.join('Annotations'),
    'JPEGImages': os.path.join('JPEGImages')
}
```

```python
string_base = "BloodImage_"
```

```python
def copy_images_and_annotations(current_object, name):
    annotation_path = os.path.join(paths['Annotations'], current_object+'.XML')
    images_path = os.path.join(paths['JPEGImages'], current_object+'.JPG')

    if os.path.exists(annotation_path):
        shutil.copy(annotation_path, os.path.join(paths[name], current_object + '.XML'))
        print("Moving " + current_object , "to", paths[name])
    if os.path.exists(images_path):
        shutil.copy(images_path, os.path.join(paths[name], current_object + '.JPG'))
        print("Moving " + current_object , "to", paths[name])
    else:
        print("There is no", current_object, "in the directory")
```

```python
for count in index.flatten():
    current_object = string_base + str(count).zfill(5)
    multiple_flag = False
    if len(np.array([l.split(',') for l in name[10]], dtype=object)) > 1:
        multiple_flag = True
        multiple_objects = np.array([l.split(',') for l in name[10]], dtype=object)
        if multiple_objects[1][0:1] == ' ':
            multiple_objects[1] =  multiple_objects[1][1:]
        
    print(name[count][0])
    if name[count][0] == 'EOSINOPHIL':
        copy_images_and_annotations(current_object, 'EOSINOPHIL')
    elif name[count][0] == 'NEUTROPHIL':
        copy_images_and_annotations(current_object, 'NEUTROPHIL')
    elif name[count][0] == 'MONOCYTE':
        copy_images_and_annotations(current_object, 'MONOCYTE')
    elif name[count][0] == 'LYMPHOCYTE':
        copy_images_and_annotations(current_object, 'LYMPHOCYTE')
    elif name[count][0] == np.nan:
        continue
    elif multiple_flag:
        for _obj in multiple_objects: 
            copy_images_and_annotations(current_object, _obj)
```
